## Insights

1. **Bill Amounts:**
   - Range: $3 to $50, average: $20.
   - 50% bills are <$18.

2. **Tip Amounts:**
   - Range: $1 to $10.
   - Mean and median ≈ $3.

3. **Group Size:**
   - 1 to 6 people.

4. **Payments:**
   - 50% pay <$20 (bill).
   - 50% pay <$3 (tip).

5. **Linear Relationship:**
   - Direct correlation between `total_bill` and `tip`.

6. **Dependency Factors:**
   - Linear correlation for gender, day, time.
   - Stronger correlation for non-smokers.

7. **Group Size Impact:**
   - Relationship constant with group size increase.

8. **Order Frequency:**
   - Higher on weekends, dinner > lunch.

9. **Weekend Trends:**
   - Median bill higher on Sat/Sun.

10. **Tip Trends:**
    - Median tip higher during dinner.

11. **Gender Distribution:**
    - ~160 males, ~80 females.
    - More males on weekends.

12. **Tip Gender Dynamics:**
    - Similar median tips for both.
    - More male outliers (higher tips).

13. **Smoking Habits:**
    - Non-smoker tables ≈ 60 more.
    - More non-smokers on Thu/Sun.

14. **Smoking and Tips:**
    - No significant relation.
